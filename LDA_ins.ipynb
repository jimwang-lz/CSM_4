```python
import numpy as np
import pandas as pd
import codecs
import collections
import numpy as np
import lda
```

```python

cutWordsFile = 'instagram_1.csv'   
wordSet = set()
for eachLine1 in codecs.open(cutWordsFile, 'r', 'utf-8'):
    lineList1 = eachLine1.split(' ')
    for i in range(len(lineList1)):
        wordSet.add(lineList1[i].strip())
wordList = list(wordSet)
```

```python

wordMatrix = []
for eachLine2 in codecs.open(cutWordsFile, 'r', 'utf-8'):
    docWords = eachLine2.strip().split(' ')
    dict1 = collections.Counter(docWords)
    key1 = list(dict1.keys())
    r1 = []
    for i in range(len(wordList)):
        if wordList[i] in key1:
            r1.append(dict1[wordList[i]])
        else:
            r1.append(0)
    wordMatrix.append(r1)
X = np.array(wordMatrix)   #词频矩阵
```

```python
#模型训练
model15 = lda.LDA(n_topics = 15, n_iter = 1500, random_state = 1)
model15.fit(X)
```

```python
#模型训练
model20 = lda.LDA(n_topics = 20, n_iter = 1500, random_state = 1)
model20.fit(X)
```

```python
#模型训练
model25 = lda.LDA(n_topics = 25, n_iter = 1500, random_state = 1)
model25.fit(X)
```

### topic classification with 3 different K

```python
#most import n words in per topic for 15 topics
n = 10
topic_word_15 = model15.topic_word_
print('==================topic top' + str(n) + ' word within 15 topics ==================')
for i, topic_dist in enumerate(topic_word_15):
    topic_words = np.array(wordList)[np.argsort(topic_dist)][:-(n+1):-1]
    print('*Topic {}\n-{}'.format(i+1, ' '.join(topic_words)))
```
